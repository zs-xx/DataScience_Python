```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
plt.plot(x, y, 'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X vals over Y')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# OO
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('X vals over Y')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes2.plot(y, x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x, y)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
plt.plot(x, y, label="y vs x")
plt.plot(y, x, label="x vs y")
plt.legend(loc=(0.5, 0.5))
```

```python
plt.plot(x, y, color='#DD8CDC', linewidth=4, alpha=0.25, linestyle='--', marker='o',
        markersize=10, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='black')
```

```python
plt.plot(x, y)
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```

```python
plt.scatter(x, y)
```

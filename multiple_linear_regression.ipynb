```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
sns.set()
```

```python
# yhat = b0 + b1x1 + b2x2 + ... + bkxk
# yhat is inferred val
# b0 is intercept
# independent vars x1 to xk

# best fitting model, min SSE = least sum of squared errs
# increase the explanatory power of the model to decrease SSE increase SSR

# adjusted r squared

data = pd.read_csv('multiple_linear_regression.csv')
# random 1,2,3
# GPA = b0 + b1SAT + b2Rand123
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Rand 1,2,3']]
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
# we increased explanatory power of the model by increasing r squared
# adjusted r increased , we were penalized , we added info but lost val
# random vars p val is 0.762 this is so high! insignificant
# y = 0.296 + 0.0017x1 - 0.0083x2

# f statistics f distribution - test overall significance of model
```
